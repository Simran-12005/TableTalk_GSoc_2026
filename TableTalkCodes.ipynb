```python
import librosa
import numpy as np
import pandas as pd
import os

DATASET_PATH = "/content/dataset"

data = []

# Extract emotion label from filename (RAVDESS)
def get_label(filename):
    try:
        return int(filename.split("-")[2])
    except:
        return -1

def extract_features(file_path):
    y, sr = librosa.load(file_path, duration=3)

    # Normalize
    y = librosa.util.normalize(y)

    # MFCC
    mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13), axis=1)

    # Pitch
    pitches, magnitudes = librosa.piptrack(y=y, sr=sr)
    pitch = np.mean(pitches[pitches > 0]) if np.any(pitches > 0) else 0

    # Energy
    energy = np.mean(librosa.feature.rms(y=y))

    # Spectral centroid
    spec_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))

    # Duration
    duration = librosa.get_duration(y=y, sr=sr)

    return list(mfcc) + [pitch, energy, spec_centroid, duration]


# 🔥 FIX: Use os.walk instead of os.listdir
count = 0
MAX_FILES = 100  # limit for speed

for root, dirs, files in os.walk(DATASET_PATH):
    for file in files:
        if file.endswith(".wav"):
            file_path = os.path.join(root, file)

            try:
                features = extract_features(file_path)
                label = get_label(file)

                data.append([file, label] + features)

                count += 1
                if count >= MAX_FILES:
                    break

            except Exception as e:
                print("Error processing:", file, e)

    if count >= MAX_FILES:
        break


# Columns
columns = ["filename", "label"] + [f"mfcc_{i}" for i in range(13)] + [
    "pitch", "energy", "spectral_centroid", "duration"
]

# DataFrame
df = pd.DataFrame(data, columns=columns)

# Save CSV
df.to_csv("/content/features.csv", index=False)

print("Task 1 Completed Successfully!")
print("Processed files:", len(df))
df.head()
```

```python
import whisper
import os
import pandas as pd

# Load model
model = whisper.load_model("base")

transcriptions = []

count = 0
MAX_FILES = 5  # small subset

for root, dirs, files in os.walk("/content/dataset"):
    for file in files:
        if file.endswith(".wav"):
            path = os.path.join(root, file)

            result = model.transcribe(path)

            text = result["text"].strip()

            transcriptions.append([file, text])

            print(f"{file} → {text}")

            count += 1
            if count >= MAX_FILES:
                break
    if count >= MAX_FILES:
        break

# Save output
df_trans = pd.DataFrame(transcriptions, columns=["filename", "transcription"])
df_trans.to_csv("transcriptions.csv", index=False)

print(" Transcription completed!")
df_trans.head()
```

```python
ground_truth = [
    "Kids are talking by the door",
    "Dogs are sitting by the door",
    "The guards are sitting by the door"
]
```

```python
def simple_accuracy(pred, actual):
    pred_words = pred.lower().split()
    actual_words = actual.lower().split()

    matches = sum(1 for w in pred_words if w in actual_words)
    return matches / len(actual_words)

# Compare first few samples
for i in range(min(len(df_trans), len(ground_truth))):
    pred = df_trans.loc[i, "transcription"]
    actual = ground_truth[i]

    acc = simple_accuracy(pred, actual)

    print(f"\nFile: {df_trans.loc[i, 'filename']}")
    print(f"Predicted: {pred}")
    print(f"Actual: {actual}")
    print(f"Accuracy: {acc:.2f}")
```

```python
import pandas as pd

df = pd.read_csv("/content/features.csv")
df.head()
```

```python
emotion_map = {
    1: "neutral",
    2: "calm",
    3: "happy",
    4: "sad",
    5: "angry",
    6: "fearful",
    7: "disgust",
    8: "surprised"
}

df["emotion"] = df["label"].map(emotion_map)
```

```python
def get_calm_long(df):
    return df[(df["emotion"] == "calm") & (df["duration"] > 2)]
```

```python
def get_high_energy(df):
    return df[df["energy"] > df["energy"].mean()]
```

```python
def get_angry(df):
    return df[df["emotion"] == "angry"]
```

```python
print("Calm narration:")
print(get_calm_long(df)[["filename", "emotion", "duration"]].head())

print("\nHigh energy speech:")
print(get_high_energy(df)[["filename", "energy"]].head())

print("\nAngry dialogue:")
print(get_angry(df)[["filename", "emotion"]].head())
```

```python
def custom_query(df, emotion=None, min_energy=None, min_duration=None):
    result = df

    if emotion:
        result = result[result["emotion"] == emotion]

    if min_energy:
        result = result[result["energy"] >= min_energy]

    if min_duration:
        result = result[result["duration"] >= min_duration]

    return result
```
